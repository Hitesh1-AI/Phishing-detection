```python
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
# load data
dataset = pd.read_csv("emails.csv")
# split data into X and Y
X = dataset.iloc[:,1:57]
y = dataset.iloc[:, 57]
class_names = dataset.iloc[:, 57]
#%%
X
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators = 100)
model.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
CV = 10
```

```python
Acc_scores = cross_val_score(model, X, y, cv = CV)
print("Accuracy: %0.3f (+/- %0.3f)" % (Acc_scores.mean(), Acc_scores.std() * 2))
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
#Print the Confusion Matrix
target_names = ['NotSPAM','SPAM' ]
plt.figure(figsize = (6, 4))
sns.heatmap(cm,cmap = 'coolwarm',linecolor = 'white',linewidths = 1,
 xticklabels = target_names,
 yticklabels = target_names,
 annot = True,
 fmt = 'd')
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```
